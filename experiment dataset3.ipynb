```python
import pandas as pd

studentInfo = pd.read_csv("dict_stud.csv")
studentInfo
```

```python
df_clicks = pd.read_csv("40weeks.csv")
df_clicks
```

```python
df_student = df_clicks.merge(studentInfo, on=['id_student', 'code_module', 'code_presentation'])
df_student
```

```python
df_student = df_student[(df_student["final_result"] == "Pass") | (df_student["final_result"] == "Fail")]
```

```python
df_student
```

```python
temp = df_student.groupby(['id_student','code_presentation', 'code_module'], as_index=False).cumcount()
```

```python
L = (df_student.set_index(['id_student','code_presentation', 'code_module',temp])
       .groupby(level=0)
       .apply(lambda x: x.values.tolist())
       .tolist())
len(L[0])
```

```python
for i in range(len(L)):
    if(len(L[i]) != 40):
        print(i)
```
