```python
import pandas as pd #import pandas
import csv #import csv module
c = pd.read_csv('studentVle.csv') #read students interactions with vle
```

```python
df = c
vle_sites = pd.read_csv("vle.csv") #read vle interactions
df_activity = df.merge(vle_sites[['id_site', 'activity_type']], on='id_site') #merge the students info with vle to get the activity type
df_activity = df_activity.drop(columns=['id_site']) # delete column id site.
df_activity.head()#print
```

```python
df_activity['date'] = df_activity['date']//7 # change date to the number of the week
df_activity['date'] = df_activity['date'].mask(df_activity['date'].lt(0),0) #remove negative values, round them
df_activity.head()# print
```

```python
new_df = df_activity.groupby(['code_module','code_presentation', 'id_student', 'date', 'activity_type'],as_index = False).sum() 
#group all row with the same features and sum the number of clicks
new_df.head()#print
```

```python
new_df.activity_type.unique() #prin all unique values of activity types.
```

```python
table = pd.pivot_table(new_df, values='sum_click', index=['code_module','code_presentation', 'id_student', 'date'], columns=['activity_type']).reset_index()
#transform all the activity type in columns and assign specific value
table = table.fillna(0) #fill nan values with 0
table.head() #print
```

```python
import numpy as np #import numpy
max_length_course = table.date.max() + 1 
test = (table.pivot_table(index=['code_module','code_presentation', 'id_student'],
               columns='date')
   .stack(level=0)
   .reindex(np.arange(40), axis=1)
   .fillna(0)
   .unstack(level=-1)
   .stack(level=0)
   .reset_index()
) # group by and create new records for the weeks when students dose not have any interaction
```

```python
assesments = pd.read_csv("assessments.csv") #read csv file assesments
assesments = assesments[assesments["assessment_type"] != 'Exam'] #remove the exam type
assesments
```

```python
studentassessment = pd.read_csv("studentAssessment.csv") #read students assesments
studentassessment = studentassessment.merge(assesments, on=['id_assessment']) #merge the students assesment data with assessment
studentassessment['final_grade'] = studentassessment['weight'] * studentassessment['score']/100 #calculate the grade.
studentassessment.head() #print
```

```python
studentassessment['date_submitted'] = studentassessment['date_submitted']//7 # get the week number
studentassessment['date_submitted'] = studentassessment['date_submitted'].mask(studentassessment['date_submitted'].lt(0),0) #remove negatives and round
studentassessment.head() #print
```

```python
studentassessment = studentassessment.drop(columns=['id_assessment','date', 'is_banked', 'score','assessment_type', 'weight']) #drop columns
studentassessment = studentassessment.rename(columns={ "date_submitted" : "date"}) #rename column date_submitted
studentassessment = studentassessment.groupby(['id_student', 'code_presentation', 'code_module', 'date'],as_index = False).sum() # group by and sum the grade
studentassessment.head() #print
```

```python
asd  = test.merge(studentassessment, on=['id_student', 'code_module', 'code_presentation', 'date'], how='left') #merge interaction table with grades
asd = asd.fillna(0) #fill nan
asd.head() #print
```

```python
asd["id"] = asd['id_student'].astype(str) + asd['code_module'] + asd['code_presentation']
asd=asd.drop(columns=['code_module','code_presentation', 'id_student'])
asd.head()
```

```python
#final  = asd.set_index(['id', 'date']).groupby(level=0).cumsum().reset_index() #cumulative sum of the table
#final #print
```

```python

studentInfo = pd.read_csv("dict_stud.csv") #read students information
studentInfo.head() #print
```

```python
studentInfo["id"] = studentInfo['id_student'].astype(str) + studentInfo['code_module'] + studentInfo['code_presentation'] #merge student info to create identifier
studentInfo = studentInfo.drop(columns=['code_module','code_presentation', 'id_student']) #drop columns
studentInfo.head()#print
```

```python
asd2  = asd.merge(studentInfo, on=['id'], how='left') #merge the student info with interactions and grades
asd2.head() #print
```

```python
df_student = asd2[(asd2["final_result"] == "Pass") | (asd2["final_result"] == "Fail")] #get the students that either pass or fail
df_student.head() #print
```

```python
Y= df_student[['id','final_result']] # assign the label array
X= df_student #assign the data array
```

```python
X = X.drop(columns=['date']) #drop columns
X.head()
```

```python
temp = X.groupby(['id'], as_index=False).cumcount()
temp
```

```python
L = (X.set_index(['id',temp])
       .groupby(level=0)
       .apply(lambda x: x.values.tolist())
       .tolist())#convert everything from data frame to list
```

```python
Y = df_student[['id','final_result']] #label array
```

```python
Y.shape #label shape
```

```python
#create label list
labels = []
for i in range(len(L)):
    if L[i][0][26] == "Pass":
        labels.append(1)
    else:
        labels.append(0)
    for j in range(40):
        L[i][j].pop()
print(labels) 
```

```python
L[0:10]
```

```python
labels = np.asarray(labels) #convert to numpy array
```

```python
data_train = np.asarray(L) #convert to numpy array
```

```python
data_train1 = data_train[:10000] #split data to training data
```

```python
labels1=labels[:10000] #split labels to training labels
```

```python
val_data = data_train[10000:13000] #validation set
val_labels = labels[10000:13000] #validation labels
test_data1 = data_train[13000:] #testing set
test_labels1 = labels[13000:] #testing labels
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding
from tensorflow.keras.layers import LSTM
from tensorflow.keras.metrics import Precision, Recall
from tensorflow.keras.callbacks import TensorBoard
tensorboard = TensorBoard(log_dir='C:\logs1\{}'.format("FINAL"))
model = Sequential()
model.add(LSTM(128,dropout=0.05, return_sequences=True, input_shape=(None, 26)))
model.add(LSTM(64))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# try using different optimizers and different optimizer configs
model.compile(loss='binary_crossentropy',
              optimizer='adam', metrics=['acc', Precision(), Recall()])

print('Train...')
history = model.fit(data_train1,labels1, validation_data=(val_data, val_labels), batch_size=32,epochs=15, callbacks=[tensorboard])
    
```

```python
import matplotlib.pyplot as plt

history_dict = history.history
loss = history_dict['acc']
val_loss = history_dict['val_acc']

epochs = range(1, len(loss) + 1)

blue_dots = 'bo'
solid_blue_line = 'b'

plt.plot(epochs, loss, blue_dots, label = 'Training Accuracy')
plt.plot(epochs, val_loss, solid_blue_line, label = 'Validation Accuracy')
plt.title("Training and validation Accuracy")
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.savefig('overfitting.eps', format = 'eps')
plt.show()
```

```python
test_accuracy = model.evaluate(test_data1, test_labels1)
test_accuracy
```
