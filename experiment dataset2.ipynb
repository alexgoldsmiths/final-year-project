```python
students = pd.read_csv('studentInfo.csv')
students = students.drop(columns=['region', 'imd_band', 'studied_credits'])
students.to_csv("stud.csv", index=False)
students
```

```python
df_student = final.merge(students, on=['id_student', 'code_module', 'code_presentation'])
df_student
```

```python
df_student = df_student.drop(columns=['region', 'imd_band', 'studied_credits'])
df_student
```

```python
df_student['gender'] = pd.Categorical(df_student['gender'])
df_student['gender'] = df_student.gender.cat.codes #encode code_presentation
df_student
```

```python
df_student['age_band'] = pd.Categorical(df_student['age_band'])
df_student['age_band'] = df_student.age_band.cat.codes #encode code_presentation
df_student
```

```python
df_student['disability'] = pd.Categorical(df_student['disability'])
df_student['disability'] = df_student.disability.cat.codes #encode code_presentation
df_student
```

```python
df_student['highest_education'] = pd.Categorical(df_student['highest_education'])
df_student['highest_education'] = df_student.highest_education.cat.codes #encode code_presentation
df_student
```

```python
df_student = df_student[(df_student['final_result'] == "Pass") | (df_student['final_result'] == "Fail")]
```

```python
num_weeks = df_student['date'].max()
```

```python
list1=[]
for i in range(10):
    list2=[]
    for j in range(num_weeks):
        row = df_student.iloc[i+j]
        if(row['date'] == j):
            list2.append(row.tolist())
        else:
            row = df_student.iloc[i+j-1]
            row['date'] = row['date'] + 1
            list2.append(row.tolist())
    i = i+num_weeks
    list1.append(list2)
print(list1)
```
